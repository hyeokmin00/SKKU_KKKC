```python
from moviepy.editor import VideoFileClip

# MP4 파일 경로
mp4_file_path = 'audio_1.mp4'

# WAV 파일 경로
wav_file_path = 'audio_1.wav'

# 비디오 파일을 로드
video_clip = VideoFileClip(mp4_file_path)

# 비디오의 오디오 추출
audio_clip = video_clip.audio

# 오디오를 WAV로 저장
audio_clip.write_audiofile(wav_file_path)

# 오디오 클립을 닫음
audio_clip.close()

print(f'{mp4_file_path}를 {wav_file_path}로 변환하였습니다.')
```

```python
import speech_recognition as sr

r = sr.Recognizer()

# 음성 파일 경로
file_path = 'audio_1.wav'

with sr.AudioFile(file_path) as source:
    audio = r.record(source)  # 음성 파일을 읽음
    text = r.recognize_google(audio, language='ko-KR')  # 한국어 음성 인식
    print("Transcript: " + text)
```
